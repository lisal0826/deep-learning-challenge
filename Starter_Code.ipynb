## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(['EIN', 'NAME'], axis=1)

```

```python
# Determine the number of unique values in each column of application_df
application_df.nunique()

# Encode categorical variables using pd.get_dummies()
application_df_encoded = pd.get_dummies(application_df)

# Split the preprocessed data into feature and target arrays, then into training and testing datasets
y = application_df_encoded['IS_SUCCESSFUL'].values
X = application_df_encoded.drop('IS_SUCCESSFUL', axis=1).values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Scale the training and testing data
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff = 150  # Adjusted cutoff value

application_types_to_replace = list(application_type_counts[application_type_counts < cutoff].index)

print(application_types_to_replace)


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# Look at CLASSIFICATION value counts that are greater than 1
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts_gt1 = classification_counts[classification_counts > 1]

print(classification_counts_gt1)
```

```python
# Choose a cutoff value for the classifications
cutoff = 100

# Create a list of classifications to be replaced
classifications_to_replace = list(classification_counts[classification_counts <= cutoff].index)

# Replace classifications in the DataFrame
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Check to see if the replacement was successful
classification_counts_post_binning = application_df['CLASSIFICATION'].value_counts()
print(classification_counts_post_binning)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_dummies = pd.get_dummies(application_df)

# Display the first few rows of the new DataFrame with dummy variables
print(application_df_dummies.head())
```

```python
# Define the features set 'X' by selecting all columns except 'IS_SUCCESSFUL'
X = application_df_dummies.drop('IS_SUCCESSFUL', axis=1).values

# Define the target set 'y' by selecting 'IS_SUCCESSFUL'
y = application_df_dummies['IS_SUCCESSFUL'].values


from sklearn.model_selection import train_test_split

# Splitting the data into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
import tensorflow as tf

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
y_train = y_train.astype('int')  # or `.astype('float32')` for binary classification targets
y_test = y_test.astype('int')  # or `.astype('float32')`

import numpy as np

# Check for and handle NaNs
if np.isnan(X_train).any():
    X_train = np.nan_to_num(X_train)
if np.isnan(X_test).any():
    X_test = np.nan_to_num(X_test)

# Similarly for infinite values
if np.isinf(X_train).any():
    X_train = np.clip(X_train, a_min=-np.finfo(np.float32).max, a_max=np.finfo(np.float32).max)
if np.isinf(X_test).any():
    X_test = np.clip(X_test, a_min=-np.finfo(np.float32).max, a_max=np.finfo(np.float32).max)


# Determine the number of input features
n_features = X_train.shape[1]

# Define the model - deep neural net
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=n_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))  # 'sigmoid' for binary classification

# Check the structure of the model
nn.summary()

# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
fit_model = nn.fit(X_train, y_train, epochs=100)  # epochs is the number of iterations over the entire dataset

# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

Report on the Neural Network Model for Alphabet Soup Charity Prediction
Overview of the Analysis
The purpose of this analysis was to create a machine learning model capable of predicting whether applicants to Alphabet Soup’s charity funding would be successful in using the funds effectively. Using features extracted from historical data, a deep learning model was developed and optimized with the goal of achieving a target predictive accuracy higher than 75%.

Results
Data Preprocessing
Target Variable: The target for the model was IS_SUCCESSFUL, indicating whether the funding was used effectively.
Features for the Model: The model features included variables such as APPLICATION_TYPE, AFFILIATION, CLASSIFICATION, USE_CASE, ORGANIZATION, STATUS, INCOME_AMT, SPECIAL_CONSIDERATIONS, and ASK_AMT, which represent various characteristics of the applications.
Variables Removed: The variables EIN and NAME were removed from the input data as they are identification information and do not contribute to the model's ability to predict successful use of funds.
Compiling, Training, and Evaluating the Model
Model Architecture: The neural network model consisted of an input layer, two hidden layers, and an output layer. The first hidden layer had 80 neurons, and the second hidden layer had 30 neurons, both using the ReLU activation function. The output layer used a sigmoid activation function for binary classification.
Achieving Target Performance: The final model achieved an accuracy of approximately 73.8% on the validation set, which is below the target of 75%.
Optimization Attempts: To increase model performance, several strategies were employed, including adjusting the number of neurons in the hidden layers, experimenting with different activation functions, and increasing the number of training epochs. The data was also preprocessed by scaling numerical features and encoding categorical variables.
Summary
The deep learning model developed for Alphabet Soup Charity prediction achieved a maximum accuracy of approximately 73.8%, falling slightly short of the 75% target. Despite various optimization attempts, including adjusting the model architecture and preprocessing steps, the target performance was not reached.

Recommendation for Different Model
Given the challenges in achieving the target accuracy with the current neural network model, it is recommended to explore ensemble machine learning models, such as Gradient Boosting Machines (GBM) or a Random Forest Classifier, for this classification problem. These models can handle a mix of numerical and categorical data well and often provide high accuracy and robustness against overfitting with proper tuning.

Why a Different Model Could Be Better: Ensemble models like GBM or Random Forest work by combining multiple weak learners to create a strong predictive model. They are often more interpretable and can provide insights into feature importance, helping to understand what drives successful funding applications. Additionally, these models can be less sensitive to noisy data and may perform better with the given dataset after appropriate hyperparameter tuning.
